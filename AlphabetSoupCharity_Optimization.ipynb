## Split & scale preprocessed data

```python
# Split our preprocessed data into our features and target arrays
#  YOUR CODE GOES HERE
y = applications_df_2['IS_SUCCESSFUL'].values.reshape(-1, 1)
X = applications_df_2.drop(['IS_SUCCESSFUL', 'SPECIAL_CONSIDERATIONS_N'], axis=1).values
# dropping 'SPECIAL_CONSIDERATIONS_N' bc it's equivalent to 'SPECIAL_CONSIDERATIONS_Y' & thus redundant

# Split the preprocessed data into a training and testing dataset
#  YOUR CODE GOES HERE
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
X_train_scaled.shape
```

# Compile, Train and Evaluate the Model

## Model 1; set to meet parameters from starter code

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
#  YOUR CODE GOES HERE

nn_model_1 = tf.keras.models.Sequential()

# First hidden layer
#  YOUR CODE GOES HERE
nn_model_1.add(tf.keras.layers.Dense(units=80, activation="relu", input_dim=42))

# Second hidden layer
#  YOUR CODE GOES HERE
nn_model_1.add(tf.keras.layers.Dense(units=30, activation="relu"))

# Output layer
#  YOUR CODE GOES HERE
nn_model_1.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

# Check the structure of the model
nn_model_1.summary()
```

```python
# Compile the model
#  YOUR CODE GOES HERE
nn_model_1.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
# set-up callback functions 
# found at https://www.tensorflow.org/api_docs/python/tf/keras/callbacks/ModelCheckpoint
checkpoint_filepath = '/tmp/checkpoint'
model_checkpoint_callback = tf.keras.callbacks.ModelCheckpoint(
    filepath=checkpoint_filepath, 
    save_weights_only=True,
    monitor=['acc', 'loss'],
    steps_per_execution=5
)
```

```python
# Train the model
#  YOUR CODE GOES HERE
fit_model = nn_model_1.fit(X_train_scaled, y_train, epochs=100, callbacks=[model_checkpoint_callback])
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn_model_1.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

### exporting to HDF5 file

```python
# Export our model to HDF5 file
#  YOUR CODE GOES HERE

# found this at: https://machinelearningmastery.com/save-load-keras-deep-learning-models/
# serialize model to JSON
model_json_1 = nn_model_1.to_json()
with open("AlphabetSoupCharity.json", "w") as json_file:
    json_file.write(model_json_1)
# serialize weights to HDF5
nn_model_1.save_weights("AlphabetSoupCharity.h5")
print("Saved model to disk")
 
```
