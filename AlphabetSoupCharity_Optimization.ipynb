## Split & scale preprocessed data

```python
# Split our preprocessed data into our features and target arrays
#  YOUR CODE GOES HERE
y = applications_df_2['IS_SUCCESSFUL'].values.reshape(-1, 1)
X = applications_df_2.drop(['IS_SUCCESSFUL', 'SPECIAL_CONSIDERATIONS_N'], axis=1).values
# dropping 'SPECIAL_CONSIDERATIONS_N' bc it's equivalent to 'SPECIAL_CONSIDERATIONS_Y' & thus redundant

# Split the preprocessed data into a training and testing dataset
#  YOUR CODE GOES HERE
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

# Compile, Train and Evaluate the Model

## Model 1; set to meet parameters from starter code

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
#  YOUR CODE GOES HERE

nn_model_1 = tf.keras.models.Sequential()

# First hidden layer
#  YOUR CODE GOES HERE
nn_model_1.add(tf.keras.layers.Dense(units=80, activation="relu", input_dim=42))

# Second hidden layer
#  YOUR CODE GOES HERE
nn_model_1.add(tf.keras.layers.Dense(units=30, activation="relu"))

# Output layer
#  YOUR CODE GOES HERE
nn_model_1.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

# Check the structure of the model
nn_model_1.summary()
```

```python
# Compile the model
#  YOUR CODE GOES HERE
nn_model_1.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
# set-up callback functions 
# found at https://www.tensorflow.org/api_docs/python/tf/keras/callbacks/ModelCheckpoint
checkpoint_filepath = '/tmp/checkpoint'
model_checkpoint_callback = tf.keras.callbacks.ModelCheckpoint(
    filepath=checkpoint_filepath, 
    save_weights_only=True,
    monitor=['acc', 'loss'],
    steps_per_execution=5
)
```

```python
# Train the model
#  YOUR CODE GOES HERE
fit_model = nn_model_1.fit(X_train_scaled, y_train, epochs=100, callbacks=[model_checkpoint_callback])
```

#### dealing with application types

```python
app_counts_v5 = opt_4_df['APPLICATION_TYPE'].value_counts()
app_counts_v5
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`
#  YOUR CODE GOES HERE
app_types_to_rep_5 = list(app_counts_v5[app_counts_v5 < 500].index)

# Replace in dataframe
for app in app_types_to_rep_5:
    opt_4_df['APPLICATION_TYPE'] = opt_4_df['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure binning was successful
opt_4_df['APPLICATION_TYPE'].value_counts()
```

#### dealing with classifications


```python
class_counts_v5 = opt_4_df['CLASSIFICATION'].value_counts()
class_counts_v5
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
#  YOUR CODE GOES HERE
classes_to_rep_v5 = list(class_counts_v5[class_counts_v5 < 1800].index)

# Replace in dataframe
for cls in classes_to_rep_v5:
    opt_4_df['CLASSIFICATION'] = opt_4_df['CLASSIFICATION'].replace(cls,"Other")
    
# Check to make sure binning was successful
opt_4_df['CLASSIFICATION'].value_counts()
```
